```python
from sklearn.datasets import make_regression

X, Y = make_regression(n_samples=50, n_features=1, n_informative=1, noise=10, random_state=11)
display(X, Y)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

model = LinearRegression()
model.fit(X, Y)
model.coef_, model.intercept_
coefficient = model.coef_[0]
intercept = model.intercept_
display(coefficient, intercept)
```

# Как обучается линейная регрессия
Линейная регрессия обучается на самом деле с помощью методов оптимизации. Они занимаются тем что берут какую-то функцию и находят в ней минимальное значение.(Экстрема функция)

MSE тоже функция, значит мы ее можем оптимизировать. Найти самые минимальные значения ошибки для нашей задачи.

Один из методов оптимизации это
# Градиетный спуск
**метод нахождения локального минимума с помощью движения вдоль градиента**
Програмный код градиентного спуска

```python
def func(x):
    return x**2

def gr_func(x):
    return 2 * x

# Визуальное понятие

X = np.linspace(-5, +5, 20)
Y = func(X)

plt.figure(figsize=(16, 6))
plt.plot(X, Y, 'r', label='Y(X)')
```

для нахождения минимального значения этой функции мы зададим начальную точку, откуда будем считать градиентя и скатываться в минимум

```python
start_point = 5 

X = np.linspace(-10, 10, 20)
Y = func(X)
plt.figure(figsize=(16, 6))
plt.plot(X, Y, 'r', label='Y(X)')
plt.plot(start_point, func(start_point), '-*g')
```

```python
grad = gr_func(start_point)
print(f'Градиент начальной точки: {grad}')
plt.figure(figsize=(16, 6))
plt.plot(X, Y, 'r', label='Y(X)')
plt.plot(start_point, func(start_point), '-*g')

next_point = start_point + grad
plt.plot([start_point, next_point], func(np.array([start_point, next_point])), '--g', label = 'gradient')
plt.legend()
```

**Так мы тут нашли просто градиент который показывает наискорейши рост (вверх). Нам нужно идти вниз чтобы найти минимум значения функции. Берем Антиградиент** 

```python
plt.figure(figsize=(16, 6))
plt.plot(X, Y, 'r', label='Y = func(X)')
plt.plot(start_point, func(start_point), '*-g', label='start point')
antigradient = start_point - grad 
plt.plot([start_point, antigradient], func(np.array([start_point, antigradient])), '--g', label='antigradient')
plt.legend()
```

Как видим антиградиент перескачило минимальные значения. Чтобы прийти к локальному минимуму добавим шаги обучения learning rate

```python
start_point = 5 
next_point = start_point
learnign_rate = 0.1
x=[]
x.append(next_point)
eps = 0.0001

plt.figure(figsize=(16, 6))
plt.plot(X, Y, 'r', label='Y=func(X)')

n = 100 
for i in range(n):
    current_point = next_point
    next_point = current_point - learnign_rate * gr_func(current_point)
    x.append(next_point)
    
    print(f'Итераци: {i}')
    print(f'Current point: {current_point}, next point: {next_point}')
    print('------------------------------')
    
    if (abs(current_point - next_point) <= eps):
        break
    
print(f'минимум значение функции потерь {next_point}')
X_grad = np.array(x)
plt.plot(X_grad, func(X_grad), '-*g', label = 'Antigradient')
plt.legend()
plt.show()
```

**Формула градиентного спуска это 
x_i's = current_point - learning_rate * gradient(current_point)**

# Многомерная линейная регрессия 

```python
X, y = make_regression(n_samples=50, n_features=4, n_informative=4, noise=10, random_state=11)

display(X, y)
```

```python
model = LinearRegression()
model.fit(X, y)
```

MSE равняется

```python
np.mean(df['residual'] ** 2)
```

# Своя реализация линейной регрессии 

```python
X, y, coeffs = make_regression(n_samples=50, n_features=4, n_informative=4, noise=10, coef=True, random_state=11)

display(X, y)
```

```python
coeffs
```

```python
X.shape, coeffs.shape
```

Важно чтобы количество column X совпадало с row коэф

```python
X = np.column_stack([X, np.ones((50))])
X
```

единичек добавили как скалярный вес 

```python
def mserror_mat(X, w, y):
    y_pred = X @ w
    return np.sum((y - y_pred) ** 2) / len(y_pred)

def gr_mserror_mat(X, w, y):
    y_pred = X @ w
    return 2/len(X)*(y - y_pred) @ (-X)
```

**Градиентный спуск**

```python
weights = np.zeros((X.shape[1]))
weights
```

```python
eps = 0.0001

learnign_rate = 0.1 

next_weights = weights

n = 100 
for i in range(n):
    cur_weight = next_weights
    next_weights = cur_weight - learnign_rate * gr_mserror_mat(X, cur_weight, y)
    
    print(f'Iteration: {i}')
    print(f'Current weights {cur_weight}, next weights {next_weights}')
    print(f'MSE {mserror_mat(X, cur_weight, y)}')
    print('----------------------------------------------')
    
    if np.linalg.norm(cur_weight - next_weights, ord=2) <= eps:
        break
```

```python
print('Веса при признаках')
print(f'True {coeffs}, \nTrained {next_weights[:-1]}')
```

**При обучении находятся градиенты MSE, а начальный точку модель берет случайно
