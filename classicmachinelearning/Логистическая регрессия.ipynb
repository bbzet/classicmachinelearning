```python
import pandas as pd
from sklearn.datasets import make_blobs 

X, y = make_blobs(n_samples=50, n_features=2, random_state=12, centers=2)

display(X, y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
X.mean(axis=0), X.std(axis=0)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.arange(-3, 4)
a = 1 
b = 0 

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.plot(x, a * x + b, label = f'y = {a}x + {b}')
plt.ylabel('y'); plt.xlabel('x')
plt.ylim(-3, 3); plt.xlim(-3, 3)
plt.grid()
plt.legend()
```

Тут линейная регрессия выступает в качестве Логистической регрессии. Наклон равен 1 и свободный вес равен 0 

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=25, n_features=1, n_informative =1, n_redundant=0, random_state=11, n_clusters_per_class=1, class_sep=0.4)
display(X, y)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X.mean(axis = 0), X.std(axis=0)
```

```python
plt.figure(figsize=(10, 2))
plt.scatter(X, np.zeros(X.shape[0]), c=y, s=100, edgecolors='black')
plt.plot([-0.15, -0.15], [-1, 1])
```

```python
from sklearn.linear_model import LogisticRegression

x = np.linspace(-3, 3)
model = LogisticRegression()
model.fit(X, y)
model.coef_, model.intercept_
```

```python
model_y_ask = model.coef_[0] * x + model.intercept_
model_y_ask
```

```python
plt.figure(figsize=(10, 2))
plt.plot(x, model_y_ask, linewidth=2, c='r')
plt.scatter(X, np.zeros(X.shape[0]), c=y, s=100, edgecolors='black')
plt.ylim(-1, 1), plt.xlim(-4, 4)
```

**Это указывает на точность работы. Насколько точно может предсказывать эта модель**

```python
model.score(X, y)
```

Мы сможем предсказывать модели с помощью predict и внести туда данные. 

```python
predict_y = model.predict(X)
predict_y[:5]
```

Если значение меньше 0 то оно автоматически сделает ее 0. А если больше 0 то 1. 

```python
plt.figure(figsize=(10, 2))
plt.scatter(X, predict_y, c=y, s=100, edgecolors='black')
plt.ylim(-1, 2), plt.xlim(-4, 4)
plt.ylabel('y predictions')
```

Выведем предсказывание вероятностей 

```python
pred_proba = model.predict_proba(X)
pred_proba[:5]
```

```python
pred_proba = pred_proba[:, 1]
plt.figure(figsize=(10, 2))
plt.scatter(X, pred_proba, c=y, s=100)
```

```python
def sigmoid(x):
    return 1/ (1 + np.exp(-x))

x = np.linspace(-3, 3, num=100)
model_y_ask = model.coef_[0] * x + model.intercept_

plt.figure(figsize=(10, 2))
plt.plot(x, sigmoid(model_y_ask), linewidth=2, c='r')
plt.scatter(X, predict_y, c=y, s=100)
plt.ylim(-0.2, 1.3)
plt.title('Sigmoid')
```

## Logistic Regression = Linear Regression + Sigmoid

# Правдоподобие

```python
df = pd.DataFrame(X, columns=['X'])
df['y'] = y
df['probablity'] = pred_proba
df['pred_y'] = predict_y
df
```

Умножаем все эти вероятности принадлежит ли точка к 0. 

```python
0.78 * 0.90 * 0.60 
```

```python
likelihood_1 = np.prod(df[df['y'] == 1]['probablity'])
likelihood_1
```

```python
likelihood_0 = np.prod(1 - df[df['y'] == 0]['probablity'])
likelihood_0
```

```python
likelihood = likelihood_0 * likelihood_1 
```

Это правдоподобие чтобы ее как то увеличить мы ее прологорифмируем

```python
-np.log(likelihood)/ df.shape[0]
```

А это и есть logloss
